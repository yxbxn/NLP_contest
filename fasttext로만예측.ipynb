# Fasttext로만 예측

원리를 잘 모르겠음.. 그냥 단순히 벡터 유사도 기반으로 가장 유사한 label을 찾아내는건지...

fasttexts는 label(category) 앞에  '__' __label__ __'을 붙여야함
https://www.tutorialkart.com/fasttext/train-and-test-supervised-text-classifier-using-fasttext/

```python

df2 = pd.DataFrame(columns=['category','document'])
df2['category'] = '__label__'+train_df['label']
df2['document'] = train_df['document']
print(df2)
```

```python
# fasttext 사용 용이하게 하기 위해 txt로 저장했다가 다시 불러옴
df2.to_csv('df2.txt', sep = '\t', index = False)
labeling = pd.read_csv("/content/df2.txt", sep = '\t')
```

```python
model = fasttext.train_supervised("/content/df2.txt", wordNgrams=3, epoch=15, lr=0.35,verbose=0)
```

```python
model.get_dimension
```

https://fasttext.cc/docs/en/python-module.html#train_supervised-parameters

```python
len(model.words)
```

```python
model.labels
```

```python
predictions = []
for line in test_df['document']:
    pred_label = model.predict(line, threshold=0.2)[0]
    predictions.append(pred_label)
```

```python
predictions
```

```python
anslist = []
for i in predictions:
    try:
        i = i[0].replace('_','').replace('label','')
        #print(i)
        anslist.append(i)
    except:
        anslist.append("")
```

```python
anslist[:5]
```

```python
ansdf=test_df
ansdf['label'] = anslist
```

```python
ansdf['digit_1'] = ansdf['label'].str[:1]
ansdf['digit_2'] = ansdf['label'].str[1:3]
ansdf['digit_3'] = ansdf['label'].str[3:]
ansdf[['digit_1','digit_2','digit_3']].to_csv('/content/drive/MyDrive/통계청/fasttext로만예측ver0.1.csv',encoding='utf-8')
```
